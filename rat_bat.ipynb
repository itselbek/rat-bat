```python
# ===============================
# Bat vs Rat Data Analysis Project
# ===============================

# Step 1: Mount Google Drive
from google.colab import drive
drive.mount('/content/drive')

# Step 2: Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import chi2_contingency, ttest_ind
import statsmodels.api as sm
from statsmodels.formula.api import ols
from IPython.display import display, HTML

sns.set(style="whitegrid")  # nicer plots

# Step 3: Load datasets
dataset1_path = '/content/drive/MyDrive/dataset1.csv'
dataset2_path = '/content/drive/MyDrive/dataset2.csv'

df_bats = pd.read_csv(dataset1_path)
df_rats = pd.read_csv(dataset2_path)

# Step 4: Quick overview
print("=== Overview of Bats Dataset ===")
display(df_bats.head())
print(df_bats.info())

print("\n=== Overview of Rats Dataset ===")
display(df_rats.head())
print(df_rats.info())

# Step 5: Handle missing values
df_bats.fillna(method='ffill', inplace=True)
df_rats.fillna(method='ffill', inplace=True)
print("\n✅ Missing values handled using forward fill.\n")

# Step 6: Convert time columns to datetime
df_bats['start_time'] = pd.to_datetime(df_bats['start_time'], dayfirst=True)
df_bats['rat_period_start'] = pd.to_datetime(df_bats['rat_period_start'], dayfirst=True)
df_bats['rat_period_end'] = pd.to_datetime(df_bats['rat_period_end'], dayfirst=True)
df_bats['sunset_time'] = pd.to_datetime(df_bats['sunset_time'], dayfirst=True)
df_rats['time'] = pd.to_datetime(df_rats['time'], dayfirst=True)
print("✅ Time columns converted to datetime.\n")

# ===============================
# Descriptive Statistics
# ===============================
print("=== Descriptive Statistics: Bats Dataset ===")
display(df_bats.describe())

print("\n=== Descriptive Statistics: Rats Dataset ===")
display(df_rats.describe())

# ===============================
# Visualizations
# ===============================

# 1. Bat Risk-taking
plt.figure(figsize=(8,5))
sns.countplot(data=df_bats, x='risk', palette='Set2')
plt.title('Bat Risk-taking Behavior (0 = Avoid, 1 = Risk)')
plt.show()

# 2. Risk vs Reward
plt.figure(figsize=(8,5))
sns.countplot(data=df_bats, x='risk', hue='reward', palette='Set1')
plt.title('Risk vs Reward in Bats')
plt.show()

# 3. Bat Landings per Month
plt.figure(figsize=(10,5))
sns.countplot(data=df_bats, x='month', palette='Set3')
plt.title('Bat Landings per Month')
plt.show()

# 4. Bat Landing Time by Risk
plt.figure(figsize=(10,5))
sns.boxplot(data=df_bats, x='risk', y='bat_landing_to_food', palette='Set2')
plt.title('Bat Landing to Food Time by Risk')
plt.show()

# 5. Bat Landing Time by Season
plt.figure(figsize=(10,5))
sns.boxplot(data=df_bats, x='season', y='bat_landing_to_food', palette='Set1')
plt.title('Bat Landing to Food Time by Season')
plt.show()

# 6. Distribution of Seconds After Rat Arrival
plt.figure(figsize=(10,5))
sns.histplot(df_bats['seconds_after_rat_arrival'], bins=30, kde=True, color='purple')
plt.title('Distribution of Seconds After Rat Arrival')
plt.show()

# 7. Bat Landings vs Rat Arrivals (dataset2)
plt.figure(figsize=(10,5))
sns.scatterplot(data=df_rats, x='rat_arrival_number', y='bat_landing_number', hue='month', palette='viridis')
plt.title('Bat Landings vs Rat Arrivals per 30-min period')
plt.show()

# 8. Food Availability vs Rat Minutes
plt.figure(figsize=(10,5))
sns.scatterplot(data=df_rats, x='rat_minutes', y='food_availability', hue='month', palette='coolwarm')
plt.title('Food Availability vs Rat Minutes')
plt.show()

# 9. Correlation Heatmap for Bats
plt.figure(figsize=(10,6))
sns.heatmap(df_bats.select_dtypes(include=np.number).corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap - Bats Dataset')
plt.show()

# ===============================
# Investigation A: Risk-Taking vs Rat Presence
# ===============================
# Define rat presence (threshold = 60s)
df_bats['rat_present'] = df_bats['seconds_after_rat_arrival'].apply(lambda x: 1 if x <= 60 else 0)
print("\nRat presence counts:")
display(df_bats['rat_present'].value_counts())

risk_rat = df_bats[df_bats['rat_present'] == 1]['risk']
risk_no_rat = df_bats[df_bats['rat_present'] == 0]['risk']

# T-test
t_stat, p_val = ttest_ind(risk_rat, risk_no_rat, equal_var=False)
print(f"\n=== Investigation A: Risk-Taking vs Rat Presence ===")
print(f"T-test: Risk with vs without rats -> t={t_stat:.3f}, p={p_val:.3f}")
if p_val < 0.05:
    print("✅ Significant difference in risk-taking with rats present.")
else:
    print("❌ No significant difference in risk-taking with rats present.")

# Chi-square test: Risk vs Reward
contingency_table = pd.crosstab(df_bats['risk'], df_bats['reward'])
chi2, p, dof, expected = chi2_contingency(contingency_table)
print(f"\nChi-square test: Risk vs Reward -> chi2={chi2:.3f}, p={p:.3f}")
if p < 0.05:
    print("✅ Risk-taking is significantly associated with reward.")
else:
    print("❌ No significant association between risk and reward.")

# ===============================
# Investigation B: Seasonal Effects on Bat Behavior
# ===============================
print("\n=== Investigation B: Seasonal Effects on Bat Behavior ===")
avg_risk_season = df_bats.groupby('season')['risk'].mean().reset_index()
print("Average Risk-Taking by Season:")
display(avg_risk_season)

# ANOVA
model = ols('risk ~ C(season)', data=df_bats).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print("\nANOVA table:")
display(anova_table)

p_season = anova_table['PR(>F)'][0]
if p_season < 0.05:
    print("✅ Significant difference in risk-taking across seasons.")
else:
    print("❌ No significant difference in risk-taking across seasons.")

# ===============================
# Summary Stats for Presentation
# ===============================
summary_risk = df_bats.groupby('risk').agg({
    'bat_landing_to_food':'mean',
    'seconds_after_rat_arrival':'mean',
    'hours_after_sunset':'mean'
}).reset_index()

print("\n--- Summary of Bat Risk-Taking Behaviour ---")
display(summary_risk)

summary_season = df_bats.groupby(['season','risk']).agg({
    'bat_landing_to_food':'mean',
    'seconds_after_rat_arrival':'mean'
}).reset_index()

print("\n--- Summary by Season ---")
display(summary_season)

# ===============================
# End of Analysis
# ===============================
print("\n✅ Analysis complete. All statistics and plots generated for Investigations A & B.")
```
